```python
year_sports=data.groupby("Year")["Sport"].nunique()
plt.Figure()
plt.plot(year_sports.index, year_sports.values)
plt.title("year vs no.of sports")
plt.xlabel("year")
plt.ylabel("number of sports")
plt.show()
```

```python
year_medals=data[data["Medal"] != "No medal"].groupby("Year")["Medal"].count()
plt.Figure()
plt.plot(year_medals.index, year_medals.values)
plt.title("year vs no.of medals")
plt.xlabel("year")
plt.ylabel("number of medals")
plt.show()
```

```python
# top 10 medal winning_countries.
top_countries = data[data["Medal"].notna()]
```

```python
data["Medal"].fillna("No medal",inplace=True)
```

```python
data.info()
```

```python
# sports with maximum medals.
top_sports=(data[data["Medal"]!="No medal"].groupby("Sport")["Medal"].count().sort_values(ascending=False).head(10))
plt.figure(figsize=(8,5))
sns.barplot(x=top_sports.values,y=top_sports.index)
plt.title("Top 10 sports by medal count")
plt.xlabel("medals")
plt.ylabel("sports")
plt.show()

```

```python
top_countries
```

```python
top_countries=(data[data["Medal"]!="No medal"].groupby("Team")["Medal"].count().sort_values(ascending=False).head(10))
plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.values,y=top_countries.index)
plt.title("Top countries by medal")
plt.xlabel("medals")
plt.ylabel("countries")
plt.show()
```
